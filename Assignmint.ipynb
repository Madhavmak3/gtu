```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
df = pd.read_csv('Sales.csv')
df
```

```python
fig,ax=plt.subplots(figsize=(7,3))
sns.lineplot(x='month_name',y='total_profit',data=df)
plt.show()
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as mticker


fig,ax=plt.subplots(figsize=(7,3))
arr = np.arange(1,13)
sns.lineplot(x='month_name',y='total_profit',data=df, linestyle="dotted", linewidth = 5)
plt.legend(["blue"], loc ="upper right")
sns.scatterplot(data=df, x='month_name',y='total_profit', marker="$\circ$", ec="face", s=300,color= "yellow")
ax.set_xticks(arr)
ax.set_xticklabels(arr)
ax.set_ylabel('Month Number')
ax.set_xlabel('Items solid Units number')
# for pos in['left','right','top','bottom']:plt.gca().spines[pos].set_visible(False)
plt.show()
```

```python
sumA = df['A'].sum()
sumB = df['B'].sum()
sumC = df['C'].sum()
sumD = df['D'].sum()
sumE = df['E'].sum()

lable=['A','B','C','D','E']
data=[sumA,sumB,sumC,sumD,sumE]
fig=plt.figure(figsize=(10,7))
plt.pie(data,labels= lable ,autopct='%.0f%%')
plt.show()
```
